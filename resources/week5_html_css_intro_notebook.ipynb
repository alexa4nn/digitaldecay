# Week 5: Intro to HTML + CSS  
### Building an Archive Interface

This notebook introduces the basics of HTML and CSS as tools for structuring and displaying your archive.

You will:
- learn basic HTML structure  
- style a page with CSS  
- create a simple archive webpage  

## 1. What is HTML?

HTML = structure

It defines:
- titles
- paragraphs
- images
- links

Think of HTML as describing *what things are*.

```python
html_example = '''
<html>
  <body>
    <h1>My Archive</h1>
    <p>This is a record of something I found.</p>
  </body>
</html>
'''

print(html_example)
```

## 2. Common HTML Tags

- <h1> → title  
- <p> → paragraph  
- <img> → image  
- <a> → link  

Tags describe meaning, not appearance.

```python
example_tags = '''
<h1>Title</h1>
<p>Description</p>
<img src="image.jpg">
<a href="https://example.com">Visit</a>
'''

print(example_tags)
```

## 3. Nesting (Structure)

HTML elements sit inside each other.

This creates hierarchy.

```python
nested_example = '''
<div>
  <h2>Item</h2>
  <p>Metadata about this item</p>
</div>
'''

print(nested_example)
```

## 4. What is CSS?

CSS = appearance

It controls:
- color
- size
- spacing
- layout

Think of CSS as *how things look*.

```python
css_example = '''
body {
  font-family: Arial;
}

h1 {
  color: blue;
}
'''

print(css_example)
```

## 5. Connecting HTML + CSS

```python
full_example = '''
<html>
  <head>
    <style>
      body { font-family: Arial; }
      h1 { color: red; }
    </style>
  </head>

  <body>
    <h1>Archive</h1>
    <p>Broken file</p>
  </body>
</html>
'''

print(full_example)
```

## 6. Archive Item Structure

Each item can be a "record"

```python
item_example = '''
<div class="item">
  <h3>Missing Image</h3>
  <p>Status: missing</p>
  <p>Source: unknown</p>
</div>
'''

print(item_example)
```

## 7. Styling Items

```python
style_example = '''
.item {
  border: 1px solid black;
  padding: 10px;
  margin: 10px;
}
'''

print(style_example)
```

## 8. Layout (Grid)

Use grid to organize your archive visually.

```python
grid_example = '''
.container {
  display: grid;
  grid-template-columns: repeat(3, 1fr);
  gap: 10px;
}
'''

print(grid_example)
```

## 9. Putting It Together

```python
final_page = '''
<html>
<head>
<style>
body { font-family: Arial; }

.container {
  display: grid;
  grid-template-columns: repeat(2, 1fr);
  gap: 10px;
}

.item {
  border: 1px solid black;
  padding: 10px;
}
</style>
</head>

<body>

<h1>My Archive</h1>

<div class="container">

  <div class="item">
    <h3>Item 1</h3>
    <p>Status: accessible</p>
  </div>

  <div class="item">
    <h3>Item 2</h3>
    <p>Status: missing</p>
  </div>

</div>

</body>
</html>
'''

print(final_page)
```
